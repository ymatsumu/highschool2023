```python
# run this cell before starting homework
# this cell will import all the python libraries you need for the homework

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1
# Plot y = 2x

x = np.linspace(0, 5, 100) 
y = x*2

plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = 2x')
plt.show()
```

```python
# Q2 - Sophia
# Plot y = 3x
x = np.linspace(0, 5, 15) 
y = x*3

plt.figure()
plt.plot(x, y, "#F8B195")
plt.xlabel('x')
plt.ylabel('y = 3x')
plt.show()
```
